```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
models = ("Deepwalk", "Node2vec", "MLP", "GCN", "GraphSage", "GAT", "GIN", "SGC")
edge_func = {
    'Dot Product': (0.9196, 0.9216, 0.7586, 0.8936, 0.9415, 0.9285, 0.8912, 0.9550),
    'Hadmard Product': (0.9555, 0.9567, 0.8415, 0.8867, 0.9554, 0.9494, 0.9575, 0.9830),
    'Concat': (0.9717, 0.9747, 0.7533, 0.8389, 0.8088, 0.9051, 0.9046, 0.9355),
}

x = np.arange(len(models))  # the label locations
width = 0.25  # the width of the bars
multiplier = 0

fig, ax = plt.subplots(layout='constrained')

for attribute, measurement in edge_func.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label=attribute)
    #ax.bar_label(rects, padding=3)
    multiplier += 1

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('AUC')
ax.set_title('Link prediction result')
ax.set_xticks(x + width, models)
ax.legend(loc='lower left')
# ax.set_ylim(0, 1)
fig.savefig("link_prediction.png", dpi=300)
```
